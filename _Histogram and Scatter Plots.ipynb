```python
# then learn time series from kaggle and make forecasting with linear regression, trend and seasonality 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 

inventory_data = pd.read_csv("inventory.csv", index_col= "PRODUCT_ID")
inventory_data.head()
print("run")
```

```python
plt.figure(figsize=(10,6))
plt.title("Forecasted Quantity and Lead Time")
sns.barplot(y =inventory_data["MRP_FCST_QTY"] , x = inventory_data["LEAD_TIME"])
plt.ylabel("Forecasted Quantity")
```

```python
plt.figure(figsize=(10,6))
plt.title("Product ID and Lead Time with regression line")
sns.scatterplot(y =inventory_data["LEAD_TIME"],x = inventory_data.index)
sns.regplot(y =inventory_data["LEAD_TIME"],x = inventory_data.index)
```

```python
plt.figure(figsize=(10,6))
plt.title("Forecasted Quantity of Products hued by their CCN")
sns.scatterplot(y =inventory_data["MRP_FCST_QTY"],x = inventory_data.index, hue = inventory_data["CCN"])
plt.ylabel("Forcasted Quantity")
```

```python
plt.figure(figsize= (20,20))
sns.histplot(inventory_data['DMND_WEEK_STRT_DATE'])
```
